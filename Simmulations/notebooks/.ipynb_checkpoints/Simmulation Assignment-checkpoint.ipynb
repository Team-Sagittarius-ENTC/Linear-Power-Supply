```python
# Import necessary liblaries
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as st
%matplotlib inline
```

```python
def refine(data, sampling_rate=100):
    """
    DOCSTRING: this function will be used to refine data. the data length will not be changed,
                but the sampling speed will be increased.
    """
    new_size = data.shape[0] * sampling_rate
    refined = np.zeros(shape=new_size)
    for i in range(data.size):
        refined[sampling_rate*i:sampling_rate*(i+1)] = data[i]
    return refined
```

## Q1)

```python
L = 1000 # given length of the data stream
# generate random data
bin_sequence = np.random.choice([0,1], L, p=[0.5,0.5]).astype('int8')
```

```python
# define amplitude
A = 1

# convert 0 to -A and 1 to +A
data_stream = np.copy(bin_sequence)
data_stream[data_stream == 0] = -A
data_stream[data_stream == 1] = A
```

```python
# plot generated data
sampling_speed = 100
t = np.arange(0, L * sampling_speed, 1)
refined_data = refine(data_stream, sampling_speed)

_, ax = plt.subplots(2,1, figsize=(18, 12))

ax[0].set_title("Generated Total data stream")
ax[0].plot(t, refined_data)
ax[0].set_xlabel("$Time$")
ax[0].set_ylabel("$Amplitude (Volts)$")
ax[0].grid()


ax[1].set_title("Zoomed in (for 100 pulses) data stream")
ax[1].plot(t, refined_data)
ax[1].set_xlabel("$Time$")
ax[1].set_ylabel("$Amplitude (Volts)$")
ax[1].set_xlim(0, 100 * sampling_speed)
ax[1].grid()

plt.show()
```

## Q2)

```python
# since we need to experiment with different variences in the future, 
# I decided to sript a function to generate normalized data stream (AWGN)

def generateAWGN(sig_2 = 1):
    data = np.random.normal(0, np.sqrt(sig_2), L)
    return data
```

```python
AWGN_1 = generateAWGN()
```

```python
# plot generated data
sampling_speed = 100
t = np.arange(0, L * sampling_speed, 1)
refined_data = refine(AWGN_1, sampling_speed)

_, ax = plt.subplots(2,1, figsize=(18, 12))

ax[0].set_title("Generated Total AWGN")
ax[0].plot(t, refined_data)
ax[0].set_xlabel("$Time$")
ax[0].set_ylabel("$Amplitude (Volts)$")
ax[0].grid()


ax[1].set_title("Zoomed in (for 100 pulses) AWGN")
ax[1].plot(t, refined_data)
ax[1].set_xlabel("$Time$")
ax[1].set_ylabel("$Amplitude (Volts)$")
ax[1].set_xlim(0, 100 * sampling_speed)
ax[1].grid()

plt.show()
```

## Q4)

```python
def recover(recived, sent_data, tau = 0):

    # Retreving the signal acording to the definition of Y
    Y = np.copy(recived)
    Y[Y > tau] = A
    Y[Y <= tau] = -A
    
    sent = np.copy(sent_data)
    sent[sent > 0] = A
    sent[sent <= tau] = -A
    
    # comparing with the original data

    # ALGORITHM

    # SUBTRACT original data and retrieved data
    # COUNT the number of non zero elements


    compared = np.abs(sent - Y)
    error_count = compared[compared != 0].size
    
    return error_count, round(error_count/compared.size*100,2)
```

```python
err, perc = recover(R, data_stream, 0)

print(f"\n\nNumber of Errors     = {err}")
print(f"Percentage of Errors = {perc}%\n\n")
```

```python
# Retreving the signal acording to the definition of Y
Y = np.copy(R)
Y[Y > 0] = A
Y[Y <= 0] = -A


# plot generated data
sampling_speed = 100
t = np.arange(0, L * sampling_speed, 1)

_, ax = plt.subplots(2,1, figsize=(18, 12))

ax[0].set_title("Zoomed in (for 200 pulses) Original Signal")
ax[0].plot(t, refine(data_stream, sampling_speed))
ax[0].set_xlabel("$Time$")
ax[0].set_ylabel("$Amplitude (Volts)$")
ax[0].set_xlim(0, 200 * sampling_speed)
ax[0].grid()


ax[1].set_title("Zoomed in (for 200 pulses) Retrived Signal $\tau = 0 $")
ax[1].plot(t, refine(Y, sampling_speed))
ax[1].set_xlabel("$Time$")
ax[1].set_ylabel("$Amplitude (Volts)$")
ax[1].set_xlim(0, 200 * sampling_speed)
ax[1].grid()

plt.show()
```

```python
def drawHistroGram(data, bins = 10):
    # First of we need to find limits of the histogram.
    # To find the limits, we will get the maximum of absolute values of the signal
    lim = np.ceil(np.max(np.abs(data)))
    
    # then we need to find the interval to create bins
    interval = 2*lim/bins
    
    # generate intervals
    bin_list = np.arange(-lim, lim+interval, interval)
    
    names = []
    
    # hist data
    histogram = np.zeros(shape=bins, dtype='int32')
    
    for i in range(bins):
        temp = data[data > bin_list[i]]
        temp = temp[temp < bin_list[i+1]]
        histogram[i] = temp.size
        
        names.append(f"{round(bin_list[i], 1)} - {round(bin_list[i+1],1)}")

    _, ax = plt.subplots(1,1, figsize=(18, 9))

    ax.set_title("Code generated Histogram")
    ax.set_xlabel("Amplitude Intervals")
    ax.set_ylabel("Count")
    
    ax.bar(names, histogram)
    ax.grid()
```

```python
def pltHist(data, title):
    _, ax = plt.subplots(1,1, figsize=(18, 9))

    ax.set_title(title)
    ax.set_xlabel("$Amplitude$")
    ax.set_ylabel("$Probability$")

    ax.hist(data, density=True, bins=150, label='Data')


    mn, mx = plt.xlim()
    kde_xs = np.linspace(mn, mx, 300)
    kde = st.gaussian_kde(data)

    ax.plot(kde_xs, kde.pdf(kde_xs), label="PDF", linewidth=3.0)
    ax.legend(loc="upper left")


    ax.grid()
```

### Sketchs of $f_R(r), f_R(r|S=A), f_R(r|S=A), A = 1,2,3,4$

```python
L = 100000 # given length of the data stream

# generate additive white gaussian
AWGN_1 = generateAWGN()

expected_vals = np.zeros(shape=(4,4))

As = [1,2,3,4]

fig, ax = plt.subplots(4,3, figsize=(18, 12), sharex=True, sharey=True)

fig.tight_layout()

for i in range(4):
    for j in range(1,3):
        
        # generating only A
        data_stream = np.ones(shape=L) * As[i]

        sign = ""
        if (j % 2) * -1 < 0:
            sign = "-"
            data_stream *= -1
        
        # add the noise
        R = data_stream + AWGN_1

        expected_vals[i][j] = np.average(R)
        
        ax[i][j].set_title(f"Histogram of $f(r|S={sign}A)$ $A = {As[i]}$")
        ax[3][j].set_xlabel("$Amplitude$")
        ax[i][0].set_ylabel("$Probability$")

        ax[i][j].hist(R, density=True, bins=150, label='Data')


        mn, mx = plt.xlim()
        kde_xs = np.linspace(mn, mx, 300)
        kde = st.gaussian_kde(R)

        ax[i][j].plot(kde_xs, kde.pdf(kde_xs), label="PDF", linewidth=3.0)
        ax[i][j].legend(loc="upper left")


        ax[i][j].grid()
        
for i in range(4):
    # generating only A
    bin_sequence = np.random.choice([0,1], L, p=[0.5,0.5]).astype('int8')
    # convert 0 to -A and 1 to +A
    data_stream = np.copy(bin_sequence)
    data_stream[data_stream == 0] = -As[i]
    data_stream[data_stream == 1] = As[i]

    # add the noise
    R = data_stream + AWGN_1

    _, perc = recover(R, data_stream, 0)
    
    expected_vals[i][0] = np.average(R)
    expected_vals[i][3] = perc
    
    ax[i][0].set_title(f"Histogram of $f_R(r)$ $A = {As[i]}$")
    ax[3][0].set_xlabel("$Amplitude$")
    
    ax[i][0].hist(R, density=True, bins=150, label='Data')


    mn, mx = plt.xlim()
    kde_xs = np.linspace(mn, mx, 300)
    kde = st.gaussian_kde(R)

    ax[i][0].plot(kde_xs, kde.pdf(kde_xs), label="PDF", linewidth=3.0)
    ax[i][0].legend(loc="upper left")


    ax[i][0].grid()
```

### c) Finding Expected Values

```python
print("+------------+------------------------+------------------------+------------------------+------------------------+")
print('|{:^12}|{:^24}|{:^24}|{:^24}|{:^24}|'.format("", "E(r)", "E(r | S = A)", "E(r | S = -A)", "Error Percentage"))
print("+------------+------------------------+------------------------+------------------------+------------------------+")

for i in range(4):
    a = round(expected_vals[i][0], 3)
    b = round(expected_vals[i][1], 3)
    c = round(expected_vals[i][2], 3)
    d = round(expected_vals[i][3], 3)
    print('|{:^12}|{:^24}|{:^24}|{:^24}|{:^24}|'.format(f"A = {As[i]}", a, b, c, f"{d}%"))
print("+------------+------------------------+------------------------+------------------------+------------------------+")
```

#### Observation of increasing the amplitude

<ul>
    <li> Generaly, Expected value of r given S=A approximately equal to A, and  Expected value of r given S=-A approximately equal to -A </li>
    <li> Furthermore, the expected value of R is approximately equal to 0 for all A </li>
    <li> When amplitude increases, the overlapping area of both histograms are reduced. </li>
    <li> When the overlap reduces, the reciver can clearly differentiate 0s and 1s correctly. Hence the error percentage reduces with increasing amplitude </li>
</ul>

#### Observation of amplification by facotor $\alpha$

<u> Obeservations for increasing the amplitude are still valid </u>

<ul>
    <li> Generaly, Expected value of r given S=A approximately equal to $\alpha \times A$, and  Expected value of r given S=-A approximately equal to $-\alpha \times A$ </li>
    <li> Furthermore, the expected value of R is approximately equal to 0 for all A </li>
</ul>

<u> Obeservations of amplification by facotor $\alpha$</u>

<ul>
    <li> The effect still visible around two states</li>
    <li> Since, the total amplitude of the signal is multiplied by a factor $\alpha$ the effect of noise get reduced</li>
    <li> Furthermore, the overlapping area of two states are reduced drastically. (Almost zero)</li>
    <li> Finally, the error percentage at the reciver side also get reduced </li>
</ul>
